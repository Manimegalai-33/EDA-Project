```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python

df.hist(figsize=(8,6))
plt.show()
```

```python

sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.show()
```

```python

sns.heatmap(df.drop('species', axis=1).corr(), annot=True)
plt.show()
```
